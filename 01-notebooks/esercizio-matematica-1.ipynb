### Svolgimento passo a passo

#### 1. Dati del problema

* Triangolo isoscele con base $b = 48\text{ cm}$.
* I lati obliqui sono congruenti e li indichiamo con $a$.
* Il coseno dell’angolo al vertice è

$$
\cos\widehat{A} = -\frac{7}{25}\;,
$$

quindi l’angolo è ottuso ($90^{\circ}<\widehat{A}<180^{\circ}$).

---

#### 2. Determinazione della lunghezza dei lati obliqui $a$

Applichiamo il **Teorema del Coseno** al lato di base:

$$
b^{2} \;=\; a^{2} + a^{2} - 2a^{2}\,\cos\widehat{A}
          \;=\; 2a^{2}\bigl(1-\cos\widehat{A}\bigr)
$$

Isoliamo $a^{2}$:

$$
a^{2} \;=\; \frac{b^{2}}{2\bigl(1-\cos\widehat{A}\bigr)}
           \;=\; \frac{48^{2}}{2\Bigl(1-(-\tfrac{7}{25})\Bigr)}
           \;=\; \frac{2304}{2\cdot\frac{32}{25}}
           \;=\; 2304 \cdot \frac{25}{64}
           \;=\; 900
$$

$$
\boxed{a = 30\text{ cm}}
$$

*Verifica*:

$$
\cos\widehat{A} = \frac{a^{2}+a^{2}-b^{2}}{2a^{2}}
                = \frac{2\cdot30^{2}-48^{2}}{2\cdot30^{2}}
                = \frac{1800-2304}{1800}
                = -\frac{7}{25}\quad\checkmark
$$

---

#### 3. Perimetro del triangolo

$$
P \;=\; 2a + b
     \;=\; 2\cdot30 + 48
     \;=\; 60 + 48
     \;=\; \boxed{108\text{ cm}}
$$

---

#### 4. Altri parametri utili (facoltativi)

L’altezza relativa alla base vale

$$
h = \sqrt{a^{2} - \Bigl(\frac{b}{2}\Bigr)^{2}}
    = \sqrt{30^{2} - 24^{2}}
    = 18\text{ cm}.
$$

---

#### 5. Illustrazione del risultato

Il grafico seguente mostra il triangolo (coordinate, lati e angolo ottuso); il codice è stato eseguito sopra ed è interamente riproducibile.

*(vedi figura)*


```python
import matplotlib.pyplot as plt
import math

# Coordinate dei vertici
A = (0, 0)        # vertice sinistro della base
B = (48, 0)       # vertice destro della base
C = (24, 18)      # vertice al vertice (altezza calcolata in modo che i lati obliqui misurino 30 cm)

# Per comodità di plotting
xs, ys = zip(A, B, C, A)

# Disegno
plt.figure(figsize=(6, 4))
plt.plot(xs, ys)                      # lati del triangolo
plt.scatter(xs[:-1], ys[:-1])         # vertici (escludo l'ultimo A ripetuto)

# Annotazioni delle lunghezze
plt.text((A[0] + B[0]) / 2, -1.5, "48 cm", ha="center", va="top")  # base
plt.text((A[0] + C[0]) / 2 - 1, (A[1] + C[1]) / 2 + 0.5,
         "30 cm", ha="right")                                      # lato AC
plt.text((B[0] + C[0]) / 2 + 1, (B[1] + C[1]) / 2 + 0.5,
         "30 cm", ha="left")                                       # lato BC

# Annotazione dell'angolo al vertice θ
theta_deg = math.degrees(math.acos(-7 / 25))
plt.text(C[0], C[1] + 1.5,
         rf"$\theta \approx {theta_deg:.1f}^\circ$", ha="center")

# Estetica
plt.axis("equal")
plt.axis("off")
plt.title("Triangolo isoscele con lato obliquo 30 cm e base 48 cm")

plt.show()
```
